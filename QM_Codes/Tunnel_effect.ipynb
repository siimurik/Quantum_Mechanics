```python
import time
import sympy as sy
```

```python
A, B, C, D, F, k1, k2, a, i = sy.symbols('A, B, C, D, F, k1, k2, a, i')
i = sy.sqrt(-1)
#sy.pprint(A+B)
```

```python
Mat = sy.Matrix([[1,      1,                -1,                  -1,   0                 ], 
                 [k1,   -k1,               -k2,                  k2,   0                 ],
                 [0,      0,   sy.exp(-i*k2*a),     sy.exp(-i*k2*a),   -sy.exp(-i*k1*a)  ],
                 [0,      0, k2*sy.exp(i*k2*a), -k2*sy.exp(-i*k2*a),   -k1*sy.exp(i*k1*a)]])
Mat
```

This is not good, because we have 4 equations, but 5 unknowns (A, B, C, D, F). To solve this problem, we chose one constant to have an arbitrary value. For this we set $A = 1$.

Our new system of equations looks, like this:

```python
M = sy.Matrix([[-1.,         1.        ,             -1.,                0.      ], 
               [ 1.,        -k1        ,            -k2,                 0.      ],
               [ 0.,    sy.exp(-i*k2*a),     sy.exp(-i*k2*a),    -sy.exp(-i*k1*a)],
               [ 0., k2*sy.exp( i*k2*a), -k2*sy.exp(-i*k2*a), -k1*sy.exp( i*k1*a)]])
M
```

```python
b = sy.Matrix([1., 1., 0., 0.])
y = sy.Matrix([B,   C,  D, F ])
print('{y} =')
sy.pprint(y)
print('{b} =')
sy.pprint(b)
```

Matrix form for this problem looks like
\begin{equation}
My = b \\
y = M^{-1}b 
\end{equation}

```python
print("Starting calculations:")
start = time.time()
M_inv = M.inv()
end = time.time()
print(end - start)
M_inv
```

```python
start = time.time()
M_inv_simp = sy.simplify(M_inv)
end = time.time()
print(end - start)
M_inv_simp
```

```python
x = M_inv_simp*b
sy.simplify(x)
```
